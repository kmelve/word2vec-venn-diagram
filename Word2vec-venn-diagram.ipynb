# Venn-diagrammer fra word2vec-distance

I dette dokumentet har jeg tatt ut lister fra `word2vec-distances` til en rekke ord som er relevante for mitt phd-prosjekt. Kilden her er hele det norske wikipedia. Her har jeg skrevet en kort metode for å visualisere fellestrekk i disse listene på en venn-diagramsmodell. Listene beholder sin sortering fra word2vec, hvor de øverste ordene ligger nærmest kjerneordet i distanse (for eksempel ligger «monoteistisk» nærmest «religion» og «askese» nærmest «spiritualitet». 

```python
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt
from matplotlib_venn import venn2,venn3

religion = set(['abrahamittisk','kristendom','religioner','trosretning','animisme','jødedom','polyteisme','religionen','religions','deisme','sionisme','jødedommen','protestantisme','religiøsitet','hinduisme','monoteistisk','gnostisisme','filosofier','ateisme','monoteisme','transcendent','trosretninger','synkretisme','universalisme','shaktisme','ateister','zoroastrisme','katolisisme','panteisme','konfucianisme','dualismen','dualistisk','hinduisk','religionene','polyteistisk','buddhisme','sufisme','sekulær','sjiaislam','hinduismen'])
spiritualitet = set(['åndelighet','mystisisme','troslære','forkynnelse','kabbala','retreater','mystikk','esoteriske','esoterisk','virkelighetsforståelse','spirituell','gnostisisme','meditasjon','askese','kristologi','åndelige','religiøsitet','hengivelse','panteisme','bibeltolkning','økumenikk','menneskesyn','ekklesiologi','grunnholdning','teosofi','okkultisme','meditativ','parapsykologi','deisme','åndelig','religionskritikk','spiritisme','humanistisk','åndsdåp','spirituelle','spiritualistisk','religionsdialog','okkulte','nyreligiøs','nestekjærlighet'])
åndelighet = set(['mystisisme','spiritualitet','dualistisk','vektleggingen','etikken','individualisme','mystikk','idealisme','antroposofien','askese','esoteriske','verdenssyn','rasjonalitet','menneskesyn','rasjonalisme','hedonisme','moralisme','religiøsitet','selvforståelse','panteisme','stoikerne','spirituelle','dualisme','humanismen','iboende','esoterisk','teologien','okkulte','materialisme','idealer','tenkning','tenkemåte','metafysiske','verdensbilde','vitenskapens','metafysisk','forkynnelse','nestekjærlighet','kunstens','individets'])

spirituell = set(['åndelig','spiritisme','mystisisme','spiritualitet','karismatisk','panteisme','okkult','askese','meditasjon','esoterisk','aristotelisk','spiritualistisk','åndelighet','evolusjonslære','parapsykologi','esoteriske','spirituelle','følelsenes','gnostisisme','sinnet','gjenfødelse','okkulte','hedonisme','selvet','religiøsitet','metafysisk','darwinistisk','gnostisk','åndelige','selvutvikling','kosmologi','asketisk','ekstatisk','lockes','selvforståelse','skolastiske','nyreligiøs','agnostiker','dekonstruksjon','iboende'])
åndelig = set(['åndelige','religiøs','askese','spirituell','asketisk','meditasjon','forkynnelse','frelsen','spiritualistisk','hengivelse','esoterisk','guddommelig','religiøst','meditativ','legemlig','spiritualitet','sjelens','renselse','frelse','kristen','kontemplasjon','dualistisk','apostasi','åndelighet','buddhismens','fullkommenhet','fromhet','jordisk','spirituelle','esoteriske','kristnes','troen','kirkelig','fromhetsliv','nyplatonismen','ekstatisk','helbredelse','hinduismens','personlighetens','åndens'])
religiøs = set(['åndelig','religiøst','religiøse','kristen','sekulær','gnostisk','spiritualistisk','askese','kjettersk','asketisk','puritansk','kristendom','kalvinistisk','rasjonalistisk','åndelige','gnostisisme','kristne','muslimsk','esoterisk','dualistisk','agnostiker','manikeismen','sufisme','troende','apostasi','kristnes','protestantismens','hengivelse','synkretisme','katolisisme','guddommelig','kultus','sjamanistisk','fanatisme','intellektuell','fromhet','zoroastrismen','sekulært','mystisisme','hedensk'])



def venndiagram(set1, set2, set1name, set2name):
    v = venn2([set1, set2], ('',''))

    # Set labels
    v.get_label_by_id('10').set_text('\n'.join(set1.difference(set2)))
    v.get_label_by_id('11').set_text('\n'.join(set1.intersection(set2)))
    v.get_label_by_id('01').set_text('\n'.join(set2.difference(set1)))
    
    # Hide diagrams
    v.get_patch_by_id('10').set_alpha(0.1)
    v.get_patch_by_id('11').set_alpha(0.1)
    v.get_patch_by_id('01').set_alpha(0.1)
    
    plt.annotate(set1name + ' {', xy = v.get_label_by_id('10').get_position(), xytext = (-90,-0), size = 'large',
            ha = 'center', textcoords = 'offset points')
    
    plt.annotate('} ' + set2name, xy = v.get_label_by_id('01').get_position(), xytext = (100,0), size = 'large',
            ha = 'center', textcoords = 'offset points')




    '''
    v.get_patch_by_id('10').set_size(20)
    v.get_patch_by_id('11').set_size(20)
    v.get_patch_by_id('01').set_size(20)
    '''
    matplotlib.pyplot.savefig(set1name+'-'+set2name+'.pdf')
```

```python
venndiagram(religion,spiritualitet, 'religion', 'spiritualitet')
```
